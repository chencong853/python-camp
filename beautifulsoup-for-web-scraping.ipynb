```python
from bs4 import BeautifulSoup
```

```python
URL = ' http://go.gwu.edu/souptext'
urlretrieve(URL, 'workshops.txt')
```

```python
soup = BeautifulSoup(open('workshops.txt'), "html5lib") # html5lib for??
```

```python
type(soup)
```

```python
print(soup.prettify()) # .prettify() makes the result more readable
```

```python
first_div = soup.find('div')
print(first_div)
```

```python
search_results = soup.find(id = "search_results") # id can be directly used to find certain value
print(search_results)
```

```python
events_divs = search_results.find_all('div', class_ = 'event')
print(events_divs)
```

```python
type(events_divs)
```

```python
event_rows = []

for event in events_divs:
    title = event.find(class_ = 'event-title').get_text() # ??class with underscore _
    summary = event.find(class_ = 'event-summary').get_text()
    date = event.find(class_ = 'event-date').get_text()
    time = event.find(class_ = 'event-time').get_text()
    event_rows.append([title, summary, date, time]) # a list with lists
```

```python
print(event_rows[0])
```

```python
with open("library_events_list.csv", 'w') as filehandle:
    filehandle.writelines("Titile, Summary, Date, Time \n")
    line = ''
    for i in range(len(event_rows)):
        # write each line to csv file
        for x in range(0,3):
            line = line + event_rows[i][x] + ';'
        line = line + event_rows[i][3] + '\n'
        filehandle.writelines(line)
```
